# Pyplot und Seaborn Demo

Dieses Beispiel illustriert Pyplot und Seaborn. Ziel ist es, eine grobe Idee der beiden Frameworks zu erhalten. 

Das Beispiel stammt von https://lisacharlotterost.de/2016/05/17/one-chart-code/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read data
data = pd.read_csv("../data/population_data.csv")
data.head()
```

```python
#plot chart
plt.scatter(np.log(data['income']), data['health'], s=data['population']/1000000, c='black')
plt.xlim(xmin=0) #set origin for x axis to zero
plt.show()
```

# Seaborn


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
data = pd.read_csv("../data/population_data.csv")
```

```python
#plot chart
g = sns.regplot('income', 'health', data=data, color='k',fit_reg=False)
g.set_xscale('log')
plt.show()
```
